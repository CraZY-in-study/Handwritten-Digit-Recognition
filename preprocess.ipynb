## 导入数据

```python
import numpy as np
import pandas as pd
train = pd.read_csv(r'D:\大四上\机器学习\作业 12.31\archive\mnist_train.csv')
print('训练集维度：',train.shape)
train.head()
```

```python
test = pd.read_csv(r'D:\大四上\机器学习\作业 12.31\archive\mnist_test.csv')
print('训练集维度：',test.shape)
test.head()
```

```python
train.describe()
```

```python
Xtrain=train.drop(columns=['label'],axis=1)
Ytrain=train['label']
Ytrain_flat=Ytrain.values.ravel()
Ytrain_count=np.unique(Ytrain_flat).shape[0]

Xtest=test.drop(columns=['label'],axis=1)
Ytest=test['label']
Ytest_flat=Ytest.values.ravel()
Ytest_count=np.unique(Ytest_flat).shape[0]
```

```python
def dense_to_one_hot(labels_dense,num_classes):
    num_labels = labels_dense.shape[0]
    index_offset = np.arange(num_labels) * num_classes
    labels_one_hot = np.zeros((num_labels, num_classes))
    labels_one_hot.flat[index_offset + labels_dense.ravel()] = 1
    return labels_one_hot

Ytrain_labels=dense_to_one_hot(Ytrain_flat,Ytrain_count)
Ytrain_labels=Ytrain_labels.astype(np.uint8)

Ytest_labels=dense_to_one_hot(Ytest_flat,Ytest_count)
Ytest_labels=Ytest_labels.astype(np.uint8)
```

## 对数据进行标准化

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)
```

## PCA降维

```python
from sklearn.decomposition import PCA

# 创建一个PCA对象，设置n_components为0.99，表示保留99%的方差
pca = PCA(n_components=0.99)

# 对数据进行降维
Xtrain_pca = pca.fit_transform(Xtrain_scaled)
Xtest_centered = Xtest_scaled - pca.mean_
Xtest_pca = pca.transform(Xtest_centered)

# 输出降维后的维数
print("The reduced dimension is", pca.n_components_)
```

## 将手写风格通过贝叶斯分类器分类

```python
from sklearn.mixture import GaussianMixture
from sklearn.naive_bayes import GaussianNB

# 设置n_components为5，假设有5个高斯分布
gmm = GaussianMixture(n_components=5)

# 对数据进行聚类，得到每个样本的类别标签
y = gmm.fit_predict(Xtrain_pca)

# 高斯贝叶斯分类器
gnb = GaussianNB()

# 训练分类器
gnb.fit(Xtrain_pca, y)

y_pred = gnb.predict(Xtest_pca)
```

```python
train['written_style']=y
test['written_style']=y_pred
```

```python
unique_labels = np.unique(y)
for label in unique_labels:
    subset = train[train['written_style'] == label]
    subset = subset.drop(columns=['written_style'], axis=1)
    filename = r'D:\大四上\机器学习\作业 12.31\archive\train_image_{}.csv'.format(label)
    subset.to_csv(filename, index=False)
```

```python
for label in unique_labels:
    subset = test[test['written_style'] == label]
    subset = subset.drop(columns=['written_style'], axis=1)
    filename = r'D:\大四上\机器学习\作业 12.31\archive\test_image_{}.csv'.format(label)
    subset.to_csv(filename, index=False)
```
